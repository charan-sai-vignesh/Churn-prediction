```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder,OneHotEncoder
import pickle
```

```python
data=pd.read_csv("Churn_Modelling.csv")
```

```python
data.head()
```

```python
data.columns
```

```python
data=data.drop(['RowNumber','CustomerId','Surname'],axis=1)
```

```python
data
```

```python
label_encoder=LabelEncoder()
data['Gender']=label_encoder.fit_transform(data['Gender'])
```

```python
ohe=OneHotEncoder()
geo_encoder=ohe.fit_transform(data[['Geography']])
```

```python
geo_encoder.toarray()
```

```python
ohe.get_feature_names_out(['Geography'])
```

```python
geo_encoder_df=pd.DataFrame(geo_encoder.toarray(),columns=ohe.get_feature_names_out(['Geography']))
```

```python
geo_encoder_df
```

```python
exit_index = data.columns.get_loc('Exited')
```

```python
left = data.iloc[:, :exit_index]
right = data.iloc[:, exit_index:]
```

```python
data_final = pd.concat([left,geo_encoder_df , right], axis=1)
```

```python
data_final.columns
```

```python
data_final.drop(['Geography'],axis=1,inplace=True)
```

```python
data_final
```

```python
with open('label_encoder.pkl','wb') as file:
    pickle.dump(label_encoder,file)

with open('ohe.pkl','wb') as file:
    pickle.dump(ohe,file)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
X_train
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
with open ('scaler.pkl','wb') as file:
    pickle.dump(scaler,file)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Input
from tensorflow.keras.callbacks import TensorBoard,EarlyStopping
import datetime
```

```python
(X_train.shape[1])
```

```python
model = Sequential([
    Input(shape=(X_train.shape[1],)),  
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
model.summary(
    
)
```

```python
model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.01),loss=keras.losses.BinaryCrossentropy(),
              metrics=[ keras.metrics.BinaryAccuracy()])
```

```python
log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
```

```python
tensorboard_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)
```

```python
early_stopping_callback=EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
```

```python
history=model.fit(
    X_train,Y_train,validation_data=(X_test,Y_test),epochs=100,
    callbacks=[tensorboard_callback,early_stopping_callback]
)
```

```python
model.save('model.keras')
```
